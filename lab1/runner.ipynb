```python
from pwn import *
import subprocess
import numpy as np
import matplotlib.pyplot as plt
```

```python
powers = [3, 5, 10, 15, 18]
x = np.array(powers)
y1 = []
y2 = []
y3 = []


for power in powers:
    p = subprocess.run(['build-debug/lab1', str(power)], stdout=subprocess.PIPE)
    t1, t2, t3, _ = p.stdout.decode().split('\n')
    y1.append(t1)
    y2.append(t2)
    y3.append(t3)

y1 = np.array(y1)
y2 = np.array(y2)
```
